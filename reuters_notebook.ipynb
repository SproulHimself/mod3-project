```python
def remove_header(dirty_string):
    sep = "(Reuters) -" 
    if sep in dirty_string:
        clean = dirty_string.split(sep, 1)[1]
    else:
        clean = dirty_string
    return clean
```

```python
def remove_footer(dirty_string):
    sep = " Reporting by "
    if sep in dirty_string:
        clean = dirty_string.split(sep, 1)[0]
    else:
        clean = dirty_string
    return clean
```

```python
clean_reuters = []

for i in range(0, len(reuters_bods)):
    x = remove_header(reuters_bods[i])
    clean_reuters.append(x)
```

```python
final_clean_reuters = []

for i in range(0, len(clean_reuters)):
    x = remove_footer(clean_reuters[i])
    final_clean_reuters.append(x)
```

```python
def process_article(article):
    tokens = word_tokenize(article)
    token_list = [word.lower() for word in tokens if word not in sw_list]
    return token_list
```
